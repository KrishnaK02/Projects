```python
# Here we are trying to fetch total runs a player has made against the team.

player = runs[runs['batsman']=='SK Raina']
player.groupby('bowling_team')['batsman_runs'].sum().sort_values(ascending=False)
```

```python
# let's make a function for this

def player_runs(batsman_name):
    player = runs[runs['batsman']==batsman_name]
    return player.groupby('bowling_team')['batsman_runs'].sum().sort_values(ascending=False).index[0]
```

```python
player_runs('V Kohli')
```

```python
player_runs('DA Warner')
```

```python
# isin function

#find the death over for players who have played more than 200 balls. (from over 16-20)

# Strike rate = (runs scored / balls faced) × 100

death_over = runs[runs['over']>15]

death_over_batsman = death_over.groupby('batsman')['batsman_runs'].count()
death_over_200 = death_over_batsman >200
list = death_over_batsman[death_over_200].index.tolist()
```

```python
# In order to calculate strikerate, we need to calculate the total runs scored and balls played.

# we are fetching runs for batsman who are also in the list we have made above.

# then we are making a total sum of the runs

# then count of balls for the purpose of strikerate

players_in_list = death_over[death_over['batsman'].isin(list)]

runs_scored = players_in_list.groupby('batsman')['batsman_runs'].sum()

balls_played = players_in_list.groupby('batsman')['batsman_runs'].count()

strike_rate = (runs_scored/balls_played) * 100

strike_rate
```

```python
# Merge (Same as Join in SQL)

# Find the player who got an orange cap in every season.

# For that we need to merge both 'data & runs' dataframes.

new=runs.merge(data,left_on='match_id',right_on='id')


# Here we are merging both 'runs' (which consists of runs for each ball) and 'data' (which consists relevant details about matches).

# Since the default value for the type of merge to happen is 'inner', we don't need to change it for this case.

# Then we are giving the names of the common columns to do the merging. (left_on, right_on)
```

```python
# Now in order to get the total runs by each batsman in every season, we are going to apply groupby function on both season and batsman as well.

new.groupby(['season','batsman'])['batsman_runs'].sum()

# Sorted the values in descending order as we want to know the orange cap i.e. most runs scorer.

# When we try to see what type result are we getting, it's Series (multi-index series).

# But we want to convert that series into dataframe. For that, there's a function called 'reset_index()'

# Now in order to get the only player who scored the most in each season (i.e. year) we are adding drop_duplicates to the code.

# In drop_duplicates, we need to give a subset because we need to tell on which column the condition is to be satisfied in drop_duplicates.

# Finally, sort the year/season in ascending order.

new.groupby(['season','batsman'])['batsman_runs'].sum().sort_values(ascending=False).reset_index().drop_duplicates(subset='season',keep='first').sort_values('season')[['season','batsman']]
```

```python
# Pivot Table


mask=runs['batsman_runs']==6

six=runs[mask]

six.shape
```

```python
# Here we are fetching data of total runs scored by hitting sixes in each over by each team over the years.

# For instance, CSK has made 30 runs by hitting sixes in over 1 all over the years.

runs_by_sixes=six.pivot_table(index='over',columns='batting_team',values='batsman_runs',aggfunc='sum')
runs_by_sixes
```

```python
# Now let's visualize the table in some form.

import seaborn as sns

sns.heatmap(runs_by_sixes)
```
